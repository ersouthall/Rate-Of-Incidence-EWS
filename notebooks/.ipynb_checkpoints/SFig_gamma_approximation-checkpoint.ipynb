# Simulations of RoI

$n = Poi(\beta SI)=Poi(\lambda)$

RoI: $\lambda = \beta SI $

### Theory 
$var(\lambda) = var(\beta SI)$

- $\omega = \beta \Delta t (1-2\phi) \zeta $
- $<\omega^2> = \color{red}{\Delta t^2 (2 \gamma -\beta)^2 \frac{\gamma}{\beta}}$

$$ d \omega(t) = (\beta(1-2\phi)-\gamma) \omega dt + \beta(1-2\phi) \sqrt{\beta \phi (1-\phi) +\gamma \phi} dW_t $$

### Simulations
To estimate RoI from $n$: 
- RoI, $\lambda$, is the mean (and variance) of $n$
- For each simulation of $n$, we take the rolling mean to get the mean of $n$ to give an estimation of $\lambda$.
- Then we average over all simulations of $\lambda$ to give the mean of RoI and we calculate the variance over all simulations of $\lambda$ will give the variance of RoI
- NB: should work if take the rolling variance of $n$ instead of rolling mean (property of Poisson process)

To estimate RoI from $I$:
- RoI equals $\beta S I$
- Calculate $\lambda$ by taking the product of Gillespie outputs of $\beta S I$

# Changing $\beta$

```python
def IncidenceTheoryM1(X, tt,RUN, BurnT,
                        beta0, gamma, p, nu=0,mu=0, p_0=0):
    if (tt <BurnT or RUN=='Fix'):
        betaT = beta0
    elif (tt< ( BurnT) or RUN == 'Ext'):
        betaT = beta0*(1-p*(tt-BurnT))
    else:
        betaT = 1.3*gamma
    phi = X[0]*(betaT*(1-X[0])-gamma)
    I1 = (betaT*(1-2*X[0])-gamma)*X[1] + N*(X[0]**2)*betaT*gamma
    I2 = 2*(betaT*(1-2*X[0])-gamma)*X[2]+ 2*N*(X[0]**2)*betaT*gamma*X[1]+N*(gamma**2)*(betaT*(1-X[0])*X[0]+gamma*X[0])

    
    I3 = 3*(betaT*(1-2*X[0])-gamma)*X[3]+3*gamma*N*(X[0]**2)*betaT*X[2] + 3*N*(gamma**2)*( betaT*(1-X[0])*X[0]+gamma*X[0])*X[1]
#     Ikurt = 4*(betaT - gamma -2*betaT*X[0])*X[2] + 6*(betaT*(1-X[0])*X[0]+gamma*X[0])*X[1]
    return [phi, I1, I2, I3]
```

# SIR model

```python
N = 10000
parametersSIR = {'beta0': 0.1, 
               'gamma':0.01,
                'I0':  [(N-10)/N, (10)/N], 
               'nu': 0,
               'mu': 0,
               'p_0': 0,
               'psign': 0}

T = 500
```

```python
def PrevalenceTheorySIR(X, tt,RUN, BurnT,
                        beta0, gamma, p, nu=0,mu=0, p_0=0):
    S = -X[0]*beta0*X[1]
    I = X[0]*beta0*X[1]-gamma*X[1]
    return [S,I]
```

```python
SIRResults = openFile(filePath = [file for file in files if 'SIR' in file][0], 
                      realisations = repeats,
                     T=T,BT= 0,ST= 0,
                      beta0=parametersSIR['beta0'],
                     gamma=parametersSIR['gamma'],
                     nu=parametersSIR['nu'],
                     mu=parametersSIR['mu'],
                     p0=parametersSIR['p_0'],
                     p = p,
                     Initial=parametersSIR['I0'],
                     sign =parametersSIR['psign'],
                     TYPE='Ext',
                     functionPrev=PrevalenceTheorySIR,
                     functionInc=NCM2)
```

```python
def NCM2(phi, tt, BT, ST,T, RUN, beta0, gamma, p, sign):
    Results = np.zeros(((BT+T)*10, 2))
    Results[:,0] = beta0*phi[:,0]*phi[:,1]
    Results[:,1] = gamma*phi[:,1]
```
