```python
import sys
sys.path.append('../../functions')

import matplotlib.pyplot as plt
import glob
from output_funcs import openFile
from output_funcs import ODea, NC, NCM2
from output_funcs import PrevalenceTheoryM1,PrevalenceTheoryM2, PrevalenceTheoryM3
from plots_funcs import IncPrev_figures
from stats_funcs import variance_theory, variance_simulations
```

# Parameter Values

```python
param_all = {'repeats': 500,
            'T': 500, 
            'BT': 300,
            'p': 1/500,
            'ST': 0}
```

```python
parametersEm = {'beta0': 0.12, 
               'gamma':0.2,
                'I0': [0,0,0], 
               'nu': 0.001,
               'mu': 0,
               'p_0': 0,
               'psign': 1,
               'BW': 30,
               'N':100000}
parametersExt = {'beta0': 1, 
               'gamma':0.2,
                'I0': [0.8,0,0], 
               'nu': 0,
               'mu': 0,
               'p_0': 0,
               'psign': -1,
                'BW': 30,
                'N': 10000}
parametersVacc = {'beta0': 1, 
               'gamma':0.18,
                'I0':  [0.2, 0.8, 0,0,0], 
               'nu': 0,
               'mu': 0.02,
               'p_0': 0,
               'psign': 0,
                 'BW':30,
                 'N':10000}
```

# Files

```python
files = glob.glob('../data/dataHPC/*')
```

```python
emFiles = [file for file in files if 'emergence' in file]
extFiles = [file for file in files if '500' in file]
vaccFiles = [file for file in files if 'vacc' in file]
```

# Results

```python
EmResults = openFile(filePath = [file for file in emFiles if 'Em_' in file][0], 
                      realisations = param_all['repeats'],
                     T=param_all['T'],BT= param_all['BT'],
                     ST= param_all['ST'],
                      beta0=parametersEm['beta0'],
                     gamma=parametersEm['gamma'],
                     nu=parametersEm['nu'],
                     mu=parametersEm['mu'],
                     p0=parametersEm['p_0'],
                     p = param_all['p'],
                     Initial=parametersEm['I0'],
                     sign =parametersEm['psign'],
                     TYPE='Ext',
                     functionPrev=PrevalenceTheoryM3,
                     functionInc=NC,
                     functionOdea= ODea)
FixResults = openFile(filePath = [file for file in emFiles if 'Fix' in file][0], 
                      realisations = param_all['repeats'],
                     T=param_all['T'],BT= param_all['BT'],
                     ST= param_all['ST'],
                      beta0=parametersEm['beta0'],
                     gamma=parametersEm['gamma'],
                     nu=parametersEm['nu'],
                     mu=parametersEm['mu'],
                     p0=parametersEm['p_0'],
                     p = param_all['p'],
                     Initial=parametersEm['I0'],
                     sign =parametersEm['psign'],
                     TYPE='Fix',
                     functionPrev=PrevalenceTheoryM3,
                     functionInc=NC,
                     functionOdea= ODea)
```

```python
ExtResults = openFile(filePath = [file for file in extFiles if 'Ext' in file][0], 
                      realisations = param_all['repeats'],
                     T=param_all['T'],BT= param_all['BT'],
                     ST= param_all['ST'],
                      beta0=parametersExt['beta0'],
                     gamma=parametersExt['gamma'],
                     nu=parametersExt['nu'],
                     mu=parametersExt['mu'],
                     p0=parametersExt['p_0'],
                     p = param_all['p'],
                     Initial=parametersExt['I0'],
                     sign =parametersExt['psign'],
                     TYPE='Ext',
                     functionPrev=PrevalenceTheoryM1,
                     functionInc=NC)
FixResults = openFile(filePath = [file for file in extFiles if 'Fix' in file][0], 
                      realisations = param_all['repeats'],
                     T=param_all['T'],BT= param_all['BT'],
                     ST= param_all['ST'],
                      beta0=parametersExt['beta0'],
                     gamma=parametersExt['gamma'],
                     nu=parametersExt['nu'],
                     mu=parametersExt['mu'],
                     p0=parametersExt['p_0'],
                     p = param_all['p'],
                     Initial=parametersExt['I0'],
                     sign =parametersExt['psign'],
                     TYPE='Fix',
                     functionPrev=PrevalenceTheoryM1,
                     functionInc=NC)
```

```python
VaccResults = openFile(filePath = [file for file in vaccFiles if 'Ext' in file][0], 
                      realisations = param_all['repeats'],
                     T=param_all['T'],BT= param_all['BT'],
                     ST= param_all['ST'],
                      beta0=parametersVacc['beta0'],
                     gamma=parametersVacc['gamma'],
                     nu=parametersVacc['nu'],
                     mu=parametersVacc['mu'],
                     p0=parametersVacc['p_0'],
                     p = param_all['p'],
                     Initial=parametersVacc['I0'],
                     sign =parametersVacc['psign'],
                     TYPE='Ext',
                     functionPrev=PrevalenceTheoryM2,
                     functionInc=NCM2)
FixResults = openFile(filePath = [file for file in vaccFiles if 'Fix' in file][0], 
                      realisations = param_all['repeats'],
                     T=param_all['T'],BT= param_all['BT'],
                     ST= param_all['ST'],
                      beta0=parametersVacc['beta0'],
                     gamma=parametersVacc['gamma'],
                     nu=parametersVacc['nu'],
                     mu=parametersVacc['mu'],
                     p0=parametersVacc['p_0'],
                     p = param_all['p'],
                     Initial=parametersVacc['I0'],
                     sign =parametersVacc['psign'],
                     TYPE='Fix',
                     functionPrev=PrevalenceTheoryM2,
                     functionInc=NCM2)
```

# Plots

```python
stat = 'variance'
extension = '.pdf'
```

```python
r0_1_t = (parametersExt['psign']*parametersExt['gamma']/parametersExt['beta0']+1)/param_all['p'] + param_all['BT']
for data in ['Incidence', 'Prevalence']:
    if data == 'Incidence':
        ylim = 0.25
        figwidth = 5
    elif data == 'Prevalence':
        ylim = 1
        figwidth = 5.7
    

    fig, ax = plt.subplots(1, figsize = (figwidth,4))
   
    IncPrev_figures(results_dict=ExtResults,
                param_dict=parametersExt,
                model_no=1,
                data_type=data,
                statistical_func_simulation=variance_simulations,
                statistical_func_theory=variance_theory,
                ax=ax,
                param_all=param_all,
                fig_main_text = False,
                RUN ='Ext',
                r0_1_t=r0_1_t, 
#                text_col='#252525',
                text_col='white',
               border_col='#a6bddb')
    ax.set_ylim([0,ylim])
    ax.legend(loc = 'upper left', ncol =1, fontsize = 12)
    ax.set_ylabel('Variance', fontsize = 16)
    ax.set_xlabel('Time', fontsize=16)
    fig.tight_layout()
    fig.savefig('../figures/'+extension[1:]+'/'+stat+data+'_M1'+extension)
```

```python

for data in ['Incidence', 'Prevalence']:
    if data == 'Incidence':
        ylim = 0.25
        figwidth = 5
    elif data == 'Prevalence':
        ylim = 1
        figwidth = 5.7
    

    fig, ax = plt.subplots(1, figsize = (figwidth,4))
   
    IncPrev_figures(results_dict=VaccResults,
                param_dict=parametersVacc,
                model_no=2,
                data_type=data,
                statistical_func_simulation=variance_simulations,
                statistical_func_theory=variance_theory,
                ax=ax,
                param_all=param_all, 
                text_col='white',
               border_col='#a6bddb')
    ax.set_ylim([0,ylim])
#     ax.legend(loc = 'upper left', ncol =1, fontsize = 12)
    ax.set_ylabel('Variance', fontsize = 16)
    ax.set_xlabel('Time', fontsize=16)
    fig.tight_layout()
    fig.savefig('../figures/'+extension[1:]+'/'+stat+data+'_M2'+extension)
```

```python
r0_1_t = (parametersEm['gamma']/parametersEm['beta0']-1)/param_all['p'] + param_all['BT']

for data in ['Incidence', 'Prevalence']:
    if data == 'Incidence':
        ylim = 0.25
        figwidth = 5
        odea = True
    elif data == 'Prevalence':
        ylim = 1
        figwidth = 5.7
        odea = False

    fig, ax = plt.subplots(1, figsize = (figwidth,4))
    IncPrev_figures(results_dict=EmResults,
            param_dict=parametersEm,
            model_no=3,
            data_type= data,
            statistical_func_simulation=variance_simulations,
            statistical_func_theory=variance_theory,
            ax=ax,
            param_all=param_all,
            r0_1_t=r0_1_t, 
                text_col='white',
               border_col='#a6bddb',
            odea_theory=odea)
    ax.set_ylim([0,ylim])
    if data == 'Incidence':
        ax.legend(loc = 'upper left', ncol =1, fontsize = 12)
    
    ax.set_ylabel('Variance', fontsize = 16)
    ax.set_xlabel('Time', fontsize=16)
    fig.tight_layout()
    fig.savefig('../figures/'+extension[1:]+'/'+stat+data+'_M3'+extension)
```
