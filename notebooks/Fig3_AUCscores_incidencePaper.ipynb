```python
import sys
sys.path.append('../../functions')

import matplotlib.pyplot as plt
import glob
from plots_funcs import AUC_figure3

import matplotlib as mpl
mpl.rcParams['hatch.linewidth'] = 2.0
```

# File information
Each file in 'Data4AUC': labelled with 450 or 390 depending on end time

- Row order: RoI, I, P (Rate of Incidence, Incidence, Prevalence)
- column order: Variance, Coef. of V, Skewness, Kurtosis, AC(1)

```python
auc_data= glob.glob('../data/Data4AUC/*')
```

```python
extension = '.pdf'
```

```python
for model in ['Model1', 'Model2', 'Model3']:

    fig, ax = plt.subplots(1, figsize = (10,6))

    legend_handles = AUC_figure3(model=model, files=auc_data, ax=ax)
    ax.set_ylim([0,1])
    ax.tick_params(axis='both', which='major', labelsize=14)
    ax.set_ylabel('AUC score', size=16)
    ax.set_xlabel('Statistic', size = 16)
    fig.tight_layout()
    # ax.set_rasterized(True)

    fig.savefig('../figures/'+extension[1:]+'/'+'AUC_scores_'+model+extension, rasterized=True,dpi=300)
```

```python
fig, ax = plt.subplots(1, figsize = (5,1.1))
ax.legend(handles = legend_handles, 
         ncol = 2,
          bbox_to_anchor=(0.8, 0.6), fontsize = 16)
ax.axis('off')
fig.tight_layout()
# ax.set_rasterized(True)

fig.savefig('../figures/'+extension[1:]+'/'+'AUC_legend'+extension, bbox_inches = 'tight'
           , rasterized=True,dpi=300)
```
